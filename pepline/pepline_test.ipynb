```python
import pandas as pd

# Your exact file path
DATA_PATH = "/Users/kevinmbp/Desktop/2D_spec_dict/data/long_peptide/LLG6+_with_intensity_top_8000"

# Load the data exactly how your main script does
df = pd.read_csv(DATA_PATH, sep="\t", skiprows=1, header=None, engine="python")
df.columns = [
    "m/z A", "m/z B", "Covariance", "Partial Cov.",
    "Score", "Ranking", "intensity A", "intensity B",
]

# Check for missing values (NaNs)
missing_a = df["intensity A"].isna().sum()
missing_b = df["intensity B"].isna().sum()

print(f"Total rows in dataset: {len(df)}")
print("-" * 30)
print(f"Missing values in 'intensity A': {missing_a}")
print(f"Missing values in 'intensity B': {missing_b}")
print("-" * 30)

# If there are missing values, show the first few rows where they happen
if missing_a > 0 or missing_b > 0:
    print("Here are the first few rows with missing intensities:")
    bad_rows = df[df["intensity A"].isna() | df["intensity B"].isna()]
    print(bad_rows[["m/z A", "m/z B", "intensity A", "intensity B", "Ranking"]].head())
else:
    print("No missing values found! The issue might be coming from upstream data processing.")
```
